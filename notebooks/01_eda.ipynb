### Numerical Variables Identification

We identified the numerical variables in the dataset using their data types.

Among these variables, some represent real quantitative information, such as customer tenure, monthly charges, and customer lifetime value (CLTV). These variables are meaningful for analysis and modeling.

Other variables, such as zip code and geographic coordinates, are numeric in format but do not represent quantities and may not be relevant for churn prediction.

Finally, some numerical variables (Churn Value and Churn Score) are directly related to the churn outcome and must not be used as input features to avoid data leakage.

```python
df_eda = df.copy()

# Convert Total Charges to numeric for EDA only
df_eda["Total Charges"] = pd.to_numeric(df_eda["Total Charges"], errors="coerce")
```

```python
num_vars = ["Tenure Months", "Monthly Charges", "Total Charges", "CLTV"]

plt.figure(figsize=(14, 10))

for i, col in enumerate(num_vars, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df_eda[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(16, 4))

for i, col in enumerate(num_vars, start=1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=df_eda[col])
    plt.title(col)
    plt.ylabel("")      # remove y-axis label
    plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

## Analysis of Numerical Variables

The numerical variables were analyzed using histograms and boxplots.

The results show that none of the numerical variables follow a normal distribution.
Most variables are skewed, with values concentrated on one side of the distribution.

Tenure months reveals distinct customer lifecycle stages, with many customers being either new or long-term.
Monthly charges show different pricing levels, indicating heterogeneous service subscriptions.
Total charges are highly right-skewed, reflecting cumulative spending over time.
CLTV also varies widely, showing that customers have very different long-term values.

Boxplots highlight the presence of outliers in all numerical variables.
These outliers represent extreme but valid customer behaviors and should not be removed without careful consideration.

```python
cat_vars = [
    "Contract",
    "Internet Service",
    "Tech Support",
    "Online Security",
    "Payment Method"
]

plt.figure(figsize=(16, 12))

for i, col in enumerate(cat_vars, 1):
    plt.subplot(3, 2, i)
    sns.countplot(data=df, x=col, hue="Churn Value")
    plt.title(f"Churn by {col}")
    plt.xlabel(col)
    plt.ylabel("Number of Customers")
    plt.xticks(rotation=30)

plt.tight_layout()
plt.show()

```

## Categorical Variables and Customer Churn

The relationship between several categorical variables and customer churn was analyzed using bar plots.

The analysis reveals strong and consistent patterns.
Customers with month-to-month contracts show significantly higher churn compared to customers with long-term contracts.
Internet service type also impacts churn, with fiber optic customers churning more frequently.

Additional services such as tech support and online security are associated with lower churn rates.
This suggests that customers receiving more support and protection are more likely to stay.

Finally, payment method plays an important role.
Customers using electronic checks show higher churn, while customers with automatic payment methods are more stable.

These results indicate that contract commitment, service complexity, and payment behavior are key drivers of customer churn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_vars = ["Tenure Months", "Monthly Charges", "Total Charges", "CLTV"]

corr_matrix = df_eda[num_vars].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()

```

## Correlation Analysis

A correlation matrix was used to analyze relationships between numerical variables.

The strongest relationship is observed between tenure months and total charges.
This is expected, as customers who remain longer naturally accumulate higher total payments.

Monthly charges also show a moderate positive correlation with total charges, since higher monthly costs increase cumulative spending.

Customer lifetime value (CLTV) shows weaker correlations with other variables, indicating that it depends on several factors and not only on tenure or price.

It is important to note that correlation does not imply causation.
This analysis is used to understand variable relationships, not to identify direct causes.
